```python
import sqlite3
import json
from pathlib import Path

project_root = Path.cwd().parent
db_path = project_root / "chat_history.db"

def print_chat_history_db():
    """Print chat sessions and html_answers stored in the DB (messages + html list)."""
    print(f"üìÅ Chat History DB Path: {db_path}")
    print(f"‚úÖ Is there DB? {db_path.exists()}")

    sessions = {}

    if db_path.exists():
        conn = sqlite3.connect(str(db_path))
        cursor = conn.cursor()

        cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
        tables = cursor.fetchall()
        print(f"\nüìä Tables: {tables}")

        cursor.execute("SELECT * FROM chat_sessions;")
        rows = cursor.fetchall()

        if rows:
            print(f"\nüìù Total sessions: {len(rows)}")
            print("\n" + "="*80)
            for i, row in enumerate(rows, 1):
                print(f"\nüîπ Session {i}:")
                print(f"   ID: {row[0]}")
                sid = row[1] if len(row) > 1 else None
                print(f"   Session ID: {sid}")
                # Schema used by DB helper: (id, session_id, messages, html_answers, created_at, updated_at)
                created = row[4] if len(row) > 4 else None
                updated = row[5] if len(row) > 5 else None
                print(f"   Created At: {created}")
                print(f"   Updated At: {updated}")

                messages = json.loads(row[2]) if row[2] else []
                print(f"   Messages ({len(messages)} entries):")
                for j, msg in enumerate(messages, 1):
                    content = msg.get('content', '')
                    preview = content[:200] + "..." if len(content) > 200 else content
                    print(f"      {j}. Role: {msg.get('role')} | {preview}")

                # html_answers stored in column index 3 as JSON list
                html_list = []
                if len(row) > 3 and row[3]:
                    try:
                        html_list = json.loads(row[3])
                        if not isinstance(html_list, list):
                            html_list = [html_list]
                    except Exception:
                        html_list = [row[3]]

                if html_list:
                    print(f"\n   HTML Answers ({len(html_list)}):")
                    for k, html in enumerate(html_list, 1):
                        preview = html[:400] + "..." if len(html) > 400 else html
                        print(f"     {k}. {preview}")
                else:
                    print("   No HTML answers stored.")

                sessions[sid] = {"messages": messages, "html_answers": html_list, "created_at": created, "updated_at": updated}
        else:
            print("\n‚ö†Ô∏è  DB empty (no chat yet).")

        conn.close()
    else:
        print("\n‚ö†Ô∏è  DB file has not been created yet. Call ask_question() first.")

    return sessions
```
